**Ik koop mooie bloemen voor mijn moeder.** (I buy beautiful flowers for my mother.)

**De tulpen in de Keukenhof zijn mooi.** (The tulips in Keukenhof are beautiful.)

**Het meisje vindt de frikandel lekker.** (The girl finds the Dutch sausage tasty.)

**Koning Willem Alexander is niet arm. Hij is rijk.** (King Willem Alexander is not poor. He is rich.)

**Nederland is een klein land.** (Netherlands is a small country.)

**Au! Deze thee is heet!** (Ow! This tea is hot!)

**Het hotel heeft vijf sterren. Het is duur.** (The hotel has five starts. It is expensive.)

**De vrouw is moor, maar de man is lelijk.** (The woman is beautiful, but the man is ugly.)

**Het huis is vies, dus ze maakt het schoon.** (The house is dirty, so she makes it clean.)

**De man werkt langzaam, omdat hij negentig jaar oud is.** (The man works slow, because he is ninety years old.)

**De leearling vinft de les niet saai, maar interessant.** (The pupil finds the lesson not boring, but interesting.)

**Schiphol is een groot vliegveld.** (Schiphol is a big airport.)

**De meisjes zijn blij, want ze krijgen een tulp.** (The girls are happy, because they received a tulip.)

**De man koopt de auto niet. Hij is arm.** (The man did not buy the car. He is poor.)

**De voetballer loopt snel. Pedro loopt langzaam.** (The footballer runs fast. Pedro runs slow.)

**Marieke kijkt naar Martin en denkt: Hij is lief.** (Marieke looks at Martin and thinks: he is lovely.)

**"Nederland" betenkent "laag land".** ("Netherlands" means "low lands".)

**We eten een kroket bij de snackbar. Hij kost EUR 1,50. Dat is goedkoop.** (We ate a croquet at the snack bar. It cost EUR 1.50. That is cheap.)

**Het is winter en het is koud.** (It is winter and it is cold.)

**De Euromast in Rotterdam is een hoog gebouw.** (The Euromast in Rotterdam is a tall building.)

**De jongen vindt spruitjes vies.** (The boy finds Brussels sprouts disgusting.)

**Om 8 uur's morgen wordt het licht.** (At 8 o'clock in the morning it is light.)

**De auto rijdt hard: 150 kilometer per uur!** (The car drives fast: 150 kilometers per hour!)

**We werken tien uur per dag. Dat is zwaar** (We work 10 hours per day. That is hard.)

**Vandaag zijn de kinderen stout.** (Today, the children are naughty.)

**Ik lees een boek. Het boek is interessant.** (I read a book. It is interesting.)

**Het meisje slaapt met een teddybeer. De teddybeer is zacht.** (The girl sleeps with a teddy bear. The teddy bear is soft.)

**We eten spruitjes met aardappels en vlees.** (We eat Brussels sprouts with potatoes and meat.)

**In de avond is het donker.** (In the evening, it is dark.)

**Wacht niet! Nu eten, want de boerenkool met worst is nu warm.** (Do not wait! Eat now, because the kale and sausage is now warm.)
