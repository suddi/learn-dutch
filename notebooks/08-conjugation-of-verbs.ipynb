**Uit je neus eten.** (Eat out of your nose.)

**Ik zeg dit. Maar ik denk dat!** (I say this. But I think that!)

**Wat denk jij?** (What do you think?)

**Martin en Gianluca denken: Marieke is mooi!** (Martin and Gianluca think: Marieke is beautiful!)

**Waar werken jullie? Op Schiphol of in het ziekenhuis?** (Where do you all work? At Schipol or in the hospital?)

**Lars loopt naar de sporthal.** (Lars walked to the sports hall.)

**Wij leren Nederlands op de Dutch Summer School.** (We learn Dutch at the Dutch Summer School.)

**De dokter luistert goed wat de patiënt zegt.** (The doctor listened well to what the patient said.)

**Jullie weten het antwoord.** (You all know the answer.)

**Wat denk jij? Sorry, dat zeg ik niet.** (What do you think? Sorry, I am not saying that.)

**Dat is een goede vraag. De baas weet het antwoord niet.** (That is a good question. The boss does not know the answer.)

**We wachten tot de advocaat en de dokter antwoorden.** (We wait for the lawyer and the doctor to answer.)

**Luisteren is slecht. Werken is goed.** (Listening is bad. Work is good.)

**Zij zien het werkwoord in het boek.** (They see the verb in the book.)

**We wachten bij de halte. We horen de tram al, maar we zien de tram nog niet.** (We wait at the stop. We hear the tram already, but we do not see the tram yet.)

**Zij zeggen: ‘Krijg!’ De actie is nu.** (They say: ‘Get it!‘ The action is now.)

**Ik ga niet naar de winkel. De leraar gaat wel naar de winkel.** (I do not go to the store. The teacher will go the store.)

**De vrouwen slapen slecht. Dat is niet goed.** (The women do not sleep well. That is not good.)

**Jullie wachten tot morgen.** (You all wait until tomorrow.)

**De vraag is: "Wat denk jij?"** (The question is: "What do you think?")

**U loopt rechts. Ik rijd links.** (You walk right. I ride left.)

**De verkoper en de ober zeggen: ‘tot ziens!’. Ik antwoord: ‘doei!’** (The seller and the waiter say: ‘See you tomorrow!‘. I answer: ‘Bye!‘)

**We kijken naar een film.** (We watch a movie.)

**Jullie krijgen boerenkool met worst. Lekker!** (You all get kale with sausage. Yummy!)

**Jullie hebben twee auto’s.** (You all have two cars.)

**We kopen het huis voor 150.000 euro.** (We buy the house for 150,000 Euro.)

**Wij lopen naar het huis.** (We walk to the house.)

**De visser en de boer worden tachtig jaar. Dat is oud.** (The fisher and the farmer are to become 80 years old. That is old.)

**Gianluca en Martin drinken bier.** (Gianluca and Martin drink beer.)

**De vrouw gaat morgen naar het ziekenhuis.** (The woman will go to the hospital tomorrow.)

**Ik luister naar muziek.** (I listen to music.)

**Het meisje loopt links.** (The girl walks left.)

**De boekhouder en de directeur schrijven een rapport.** (The accountant and the company director wrote a report.)

**De studenten vragen: ‘Welke universiteit is goed?’ Zij weten het antwoord niet.** (The students ask: ‘Which university is good?‘ They do not know the answer.)

**‘Spreken’ is een synoniem van ‘Praten‘.** (‘Speaking‘ is a synonym with ‘Talking‘.)

**Jullie rijden in een auto. Ik rij op de fiets.** (You all ride in a car. I ride on the bicycle.)

**Ik loop niet naar het centrum.** (I do not walk to the center.)

**‘Praten’ is een werkwoord.** (‘To talk‘ is a verb.)

**Wij gaan om tien uur naar bed. We wachten niet op jou.** (We go to bed at ten o'clock. We will not wait for you.)

**We lopen naar het centrum.** (We walk to the center.)

**Hier rijden zes bussen en drie trams.** (Six busses and three trams ride here.)

**De jongen en het meisje eten drop. Lekker!** (The boy and the girl eat liquorice. Yummy!)

**Jullie lezen de vraag. Dan schrijven jullie het antwoord.** (You all read the question. Then you all write the answer.)
