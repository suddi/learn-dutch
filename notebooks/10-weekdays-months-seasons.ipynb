**Welke dag van de week?** (Which day of the week?)

**Welke dag komt na dinsdag? Ik weet het antwoord: woensdag.** (Which day comes after Tuesday? I know the answer: Wednesday.)

**Op dinsdag heb ik een examen. Het examen is voor mijn inburgering.** (On Tuesday, I have an exam. The exam is for my civic integration.)

**Op 14 februari is het Valentijnsdag. Dan krijgt het meisje mooie bloemen.** (On 14 February is Valentine's Day. Then the girl gets beautiful flowers.)

**Maandag, dinsdag, woensdag, donderdag, vrijdag. Welke werkdag mist hier?** (Monday, Tuesday, Wednesday, Thursday, Friday. Which workday is missing here?)

**Op zondag gaat mijn vader naar de kerk.** (On Sunday, my father goes to the church.)

**Het is vrijdag! Nog een dagje werken en dan is het weekend! Lekker!** (It is Friday! Yet one day of work and then it is the weekend! Awesome!)

**Op zaterdag hebben de leerlingen vrij. Ze gaan dan naar Orvelte met de leraar.** (The students have August free. They then go to Orvelte with the teacher.)

**In maart, in de lente, opent de Keukenhof. We zien daar mooie tulpen.** (In March, in the Spring, open Keukenhof. We see beautiful tulips over there.)

**De voetballers gaan morgen naar Amsterdam.** (The footballers go to Amsterdam tomorrow.)

**Op zondag gaan mijn vader en moeder naar de kerk.** (On Sunday, my father and mother go to the church.)

**Op 21 juni begint de zomer in Nederland.** (On 21 June, begins the summer in Netherlands.)

**Augustus, september, oktober, november, december. Welke maand mist hier?** (August, September, October, November, December. Which month is missing here?)

**In de winter is het koud. De dagen zijn kort en de nachten lang.** (In the winter, it is cold. The days are short and the nights long.)

**De maanden september, oktober, november, december zijn in de herfst.** (The months September, October, November, December are in the autumn.)

**De herfst begint op 21 september en duurt tot 20 december.** (The autumn begins on 21 September and takes until 20 December.)

**In welk seizoen is het koud in Nederland? Het antwoord: in de winter.** In which season is it cold in Netherlands? The answer: in the winter.)

**De lente is mijn favouriete seizoen. Ik zie dan nieuwe tulpen op de Keukenhof.** (The spring is my favourite season. I see then new tulips in Keukenhof.)

**Ik ben blij! In augustus leer ik Nederlands op de Dutch Summer School!** (I am happy! In August, I learn Dutch at the Dutch Summer School!)

**In de zomer gaan veel Nederlanders op vakantie naar Frankrijk en Spanje.** (In the summer, many Dutch people go on vacation to France and Spain.)

**In de maanden juli en augustus hebben veel Nederlanders vakantie.** (In the months July and August, many Dutch people have vacation.)

**Vandaag is het donderdag. Gisteren was het woensdag.** (Today it is Thursday. Yesterday it was Wednesday.)

**Op 27 april is het Koningsdag. Dat is een feestdag in Nederland.** (On 27 April it is King's day. That is a party day in Netherlands.)

**In welke maand begint de herfst? Ik weet het antwoord: in september.** (In which month begins the autumn? I know the answer: in September.)

**Op 25 en 26 december is het Kerst in Nederland. Dat is in de winter.** (On 25 and 26 December it is Christmas in Netherlands. That is in the winter.)

**Op woensdag koopt mijn vrouw gehakt bij de slager.** (On Wednesday, I buy my wife minced meat at the butcher.)

**Vandaag eet ik vis met groente en aardappels. Lekker!** (Today, I eat fish with vegetables and potatoes. Tasty!)

**Op welke dag van de week eet jij vis? Ik eet vis op vrijdag. Lekker!** (On which day of the week, do you eat fish? I eat fish on Friday. Tasty!)

**In je agenda zie je de dagen van de week.** (In you calendar, you can see the days of the week.)

**De toerist eet een lekkere haring met uitjes.** (The tourist eat a tasty herring with outings.)

**Het is vandaag koopzondag! Ik ga met mijn moeder naar de winkels in het centrum.** (It is shoppping Sunday today! I go with my mother to the shops in the town center.)

**Oktober is in de herfst.** (October is in the autumn.)

**Vandaag werk ik niet, want het is zondag.** (Today, I do not work, because it is Sunday.)

**Het jaar heeft 365 dagen.** (The year has 365 days.)

**Maandag is de eerste werkdag. Dan begint de werkweek.** (Monday is the first work day. Then begins the workweek.)

**De maand mei is in de lente.** (The month May is in the spring.)

**Ik werk vijf dagen per week.** (I work five days per week.)

**In januari ga ik naar de Winter School in Drenthe. Leuk!** (In January, I go to the Winter School in Drenthe. Fun!)

**Martin is met Marieke. Hij vraagt: wanneer kijken we samen naar de maan?** (Martin is with Marieke. He asks: when do we look at the moon together?)

**De dagen van de week zijn: maandag, dinsdag, woensdag, donderdag, vrijdag, zaterdag en zondag.** (The days of the week are: Monday, Tuesday, Wednesday, Thursday, Friday, Saturday and Sunday.)
